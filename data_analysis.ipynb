```python
from urllib.request import urlopen

def get_data_from_url(url):
    with urlopen(url) as conn:
        return conn.readlines()
```

```python
data = get_data_from_url('https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv?accessType=DOWNLOAD')
```

```python
popular_names = dict()
for line in data[1:]:
    line_decoded = line.decode('utf-8')
    name_information = line_decoded.split(',')
    gender = name_information[1]
    etnicity = name_information[2]
    name = name_information[3].lower()
    count = name_information[4]
    
    if gender not in popular_names:
        popular_names[gender] = dict()
    else:
        if etnicity not in popular_names[gender]:
            popular_names[gender][etnicity] = dict()
        else:
            if name not in popular_names[gender][etnicity]:
                popular_names[gender][etnicity][name] = int(count)
            else:
                popular_names[gender][etnicity][name] += int(count)
            

popular_names


    
```

```python
def sort_dictionary_by_values(dictionary, reversed=False):
    return sorted(dictionary.items(), key=lambda item: item[1], reverse=reversed)
```

```python
sort_dictionary_by_values(popular_names['FEMALE']['HISPANIC'], reversed=True)
```

```python
for gender, gender_values in popular_names.items():
    print('Popular names for gender %s' % gender)
    for etnicity, etnicity_values in gender_values.items():
        print('Popular names for etnicity %s' % etnicity)
        stats = sort_dictionary_by_values(etnicity_values, reversed=True)
        print('Names are %s, %s, %s' % (stats[0], stats[1], stats[2]))
```

```python
drug_deaths = get_data_from_url('https://data.ct.gov/api/views/rybz-nyjw/rows.csv?accessType=DOWNLOAD')
```

```python
drug_names = ['cocaine', 'heroin', 'amphetamine', 'fentanyl', 'cannabis']
drug_death_stats = dict()
for line in drug_deaths:
    line = line.decode('utf-8').lower()
    
    for drug_name in drug_names:
        if drug_name in line:
            if drug_name not in drug_death_stats:
                drug_death_stats[drug_name] = 1
            else:
                drug_death_stats[drug_name] += 1
drug_death_stats
    
```

```python
sort_dictionary_by_values(drug_death_stats, reversed=True)
```
