```python
def word_count_from_file(filename):
    word_counts = dict()

    with open(filename) as file:
        for line in file:
            list_of_words = line.split(' ')

            for word in list_of_words:
                cleaned_word = word.replace('.', '').replace('?','').replace(',','')
                cleaned_word_lowercase = cleaned_word.lower()

                if cleaned_word_lowercase not in word_counts:
                    word_counts[cleaned_word_lowercase] = 1
                else:
                    word_counts[cleaned_word_lowercase] = word_counts[cleaned_word_lowercase] + 1

    sorted_counts = sorted(word_counts.items(), key=lambda item: item[1], reverse=True)
    return sorted_counts
```

```python
from urllib.request import urlopen

def get_data_from_url(url):
    with urlopen(url) as conn:
        return conn.readlines()
    
```

```python
url = 'https://raw.githubusercontent.com/alextavgen/TrumpHallucinator/master/trump-hallucinator/src/main/resources/speeches.txt'
print(get_data_from_url(url))
```

```python
url = 'https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv?accessType=DOWNLOAD'
result = get_data_from_url(url)
```

```python
popular_names = dict()

for line in result[1:]:
    splitted_line = line.decode('utf-8').split(',')
    gender = splitted_line[1]
    etnicity = splitted_line[2]
    name = splitted_line[3]
    
    if gender not in popular_names:
        popular_names[gender] = dict()
    else:
        if etnicity not in popular_names[gender]:
            popular_names[gender][etnicity] = dict()
        else:
            if name not in popular_names[gender][etnicity]:
                popular_names[gender][etnicity][name] = 1
            else:
                popular_names[gender][etnicity][name] += 1
print(popular_names)
```

```python
overdose_stats_url = 'https://data.ct.gov/api/views/rybz-nyjw/rows.csv?accessType=DOWNLOAD'
overdose_data = get_data_from_url(overdose_stats_url)
```

```python
overdose_data
```

```python
drugs_list = ['cocaine', 'fentanyl', 'oxycodone', 'amphetamine', 'heroin', 'ethanol']
stats = dict()
for line in overdose_data:
    for drug_name in drugs_list:
        if drug_name in line.decode('utf-8'):
            # add statistics
            if drug_name not in stats:
                stats[drug_name] = 1
            else:
                stats[drug_name] += 1
```
